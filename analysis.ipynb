```python
#Sex Ratio at Birth (Male  Birth Per Female) Over Time

# Create a colorful line plot
plt.figure(figsize=(6, 3))  # Adjust the figure size as needed

plt.bar(df['year'], df['Sex ratio at birth (male births per female births)'],label='Sex Ratio')
plt.xlabel('Year')
plt.ylabel('Sex Ratio (male births per female births)')
plt.title('Sex Ratio at Birth (Male  Birth Per Female) Over Time')
plt.legend()
plt.grid(True)

plt.show()
```

```python
# Columns to compare
columns_to_compare = [
    'Population ages 10-14, female (% of female population)',
    'Population ages 65-69, female (% of female population)',
    'Population ages 55-59, female (% of female population)',
    'Population ages 45-49, female (% of female population)',
    'Population ages 35-39, female (% of female population)',
    'Population ages 25-29, female (% of female population)',
    'Population ages 80 and above, female (% of female population)',
    'Population ages 70-74, female (% of female population)',
]

# Create a single chart comparing selected columns with the 'year' column
plt.figure(figsize=(8, 6))

# Loop through the selected columns and plot each one
for column in columns_to_compare:
    label = column.split(',')[0]
    plt.plot(df['year'], df[column], marker='o', label=label)

plt.xlabel('Year')
plt.ylabel('Percentage')
plt.title('Population by Age Group Over Time (Female)')
plt.legend(loc='upper right', bbox_to_anchor=(1.25, 1),fontsize='small')
plt.grid(True)

plt.show()
```
