```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
file_path = r'C:\Learnings\India_Analysis\updated.xlsx'
df = pd.read_excel(file_path)
```

```python
df.head()
```

```python
df1 = df[df['Intentional homicides, female (per 100,000 female)']>0]
```

```python
# Plotting the data
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed

# Assuming 'Year' is the x-axis and 'Intentional homicides, female (per 100,000 female)' is the y-axis
plt.plot(df1['Year'], df1['Intentional homicides, female (per 100,000 female)'], marker='o', linestyle='-')

# Adding labels and title
plt.xlabel('Year')
plt.ylabel('Intentional homicides, female (per 100,000 female)')
plt.title('Intentional Homicides for Females over the Years')

# Show the plot
plt.grid(True)
plt.show()
```

```python
df2 = df[df['Battle-related deaths (number of people)']>0]
# Plotting the data
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed

# Assuming 'Year' is the x-axis and 'Intentional homicides, female (per 100,000 female)' is the y-axis
plt.bar(df2['Year'], df2['Battle-related deaths (number of people)'])

# Adding labels and title
plt.xlabel('Year')
plt.ylabel('Battle-related deaths (number of people)')
plt.title('Battle-related deaths (number of people)')

# Show the plot
plt.grid(True)
plt.show()
```
