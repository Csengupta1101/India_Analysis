```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
import plotly.express as px 
```

```python
df = pd.read_excel(r'C:\Learnings\India_Analysis\updated.xlsx')
df.head()
```

```python
# Intentional homicides for Females - This reflects how many women were killed in every 1000 women over the course of last few decades 
# Plotting the data as a bar plot
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
df1 = df[df['Intentional homicides, female (per 100,000 female)']>0]

# Generate shades of color
colors = plt.cm.viridis(np.linspace(0.2, 1, len(df)))

# Plotting the data as a bar plot with varying colors
plt.figure(figsize=(10, 6))

# Assuming 'Year' is the x-axis and 'Intentional homicides, female (per 100,000 female)' is the y-axis
plt.bar(df1['Year'], df1['Intentional homicides, female (per 100,000 female)'], color=colors)

# Adding labels and title
plt.xlabel('Year')
plt.ylabel('Homicide Count')
plt.title('Intentional Homicides for Females per 100,000 Females')

# Show the plot
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
